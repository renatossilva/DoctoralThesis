# Carregando os dados

```python
# Data
file_name = ['Normal_20011207_12-15_full.io', 'CodeRed_20011907_12-15_full.io']
col_names = ['relatorio', 'time', 'BPG_type', 'anomaly', 'AS_peer', 'relevance_peer', 'threat_peer', 'AS_origin', 'relevance', 'relevance_avg', 'relevance_median', 'relevance_max', 'relevance_min', 'threat', 'threat_avg', 'threat_median', 'threat_max', 'threat_min', 'repetitions', 'length', 'origin_type']

df_0 = pd.read_csv(file_name[0], sep='|', names=col_names)
df_0['time'] = pd.to_datetime(df_0['time'], format='%m/%d/%y %H:%M:%S')
df_0[['IGP', 'EGP', 'Incomplete']] = pd.get_dummies(df_0['origin_type'])

df_1 = pd.read_csv(file_name[1], sep='|', names=col_names)
df_1['time'] = pd.to_datetime(df_1['time'], format='%m/%d/%y %H:%M:%S')
df_1[['IGP', 'EGP', 'Incomplete']] = pd.get_dummies(df_1['origin_type'])
```

# Criando os clusters

```python
chosen_cols =['threat','threat_avg','threat_median', 'threat_max', 'threat_min', 'relevance', 'relevance_avg', 'relevance_median', 'relevance_max', 'relevance_min', 'length', 'repetitions', 'relevance_peer', 'threat_peer', 'IGP', 'EGP', 'Incomplete']
```

```python
X_0 = df_0.sample(frac=0.5, random_state=42)
X_1 = df_1.sample(frac=0.5, random_state=55)
```

```python
for i in [4, 5 ,6]:
  cluster = AgglomerativeClustering(n_clusters=i, affinity='euclidean', linkage='ward')
  cluster.fit_predict(X_0.loc[:, chosen_cols].values)
  X_0[str(i)+'_cluster'] = cluster.labels_
```

```python
for i in [4, 5 ,6]:
  cluster = AgglomerativeClustering(n_clusters=i, affinity='euclidean', linkage='ward')
  cluster.fit_predict(X_1.loc[:, chosen_cols].values)
  X_1[str(i)+'_cluster'] = cluster.labels_
```

```python
X_0
```

```python
X_1
```

```python
X_01 = pd.concat([X_0, X_1])
```

```python
for i in [4, 5 ,6]:
  cluster = AgglomerativeClustering(n_clusters=i, affinity='euclidean', linkage='ward')
  cluster.fit_predict(X_01.loc[:, chosen_cols].values)
  X_01['01_'+str(i)+'_cluster'] = cluster.labels_
```
